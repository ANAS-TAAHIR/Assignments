 ### Fit Linear Regression Model (Training data)

 As our linear Model was

And using derivatives we transformed our model into 2 simultaneous equations

NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
import numpy as np
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
# Assuming x is a numpy array
n_rows = x_train.shape[0]  # Number of rows in x
sum_x = np.sum(x_train)  # Sum of all elements in x
squared_x = np.square(x_train)  # Square of x
sum_squared_x = np.sum(squared_x)  # Sum of square of x
sum_y = np.sum(y_train)  # Sum of all elements in y
sum_xy = np.dot(x_train, y_train)  # Dot product of x and y
matrixA = np.array([[n_rows, sum_x], [sum_x, sum_squared_x]])  # Matrix of x
print('Matrix A:\n', matrixA)
matrixB = np.array([sum_y, sum_xy])  # Vector of y
print('Matrix B:\n', matrixB)
```

Expected output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20Linear.png?raw=true)


#### Compute Valus of Both `s


```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node and 0_1
inverse_matrixA = np.linalg.inv(matrixA)  # Inverse of matrixA
coefficients = np.dot(inverse_matrixA, matrixB)  # Coefficients of the line
print('Coefficients:\n', coefficients)
```

Expected Output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20LinearT.png?raw=true)


### Run Predictions (Testing data)

```python
# Your code Here
# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x
# _____DO NOT USE FOR LOOP_____
# instead Convert the column of test_x into numpy a 2 dimensional matrix
# then use matrix multiplication to calculate the predicted values of y
testData = pd.read_csv('testRegression.csv')
x_test = np.array(testData['X'])
y_test = np.array(testData['R'])
y_pred = coefficients[0] + coefficients[1] * x_test
print('Predicted values of y:\n', y_pred)
```

### Mean Square Error

```python
#Your code Here
#Using cost function calculate the cost of the model
cost = np.sum(np.square(y_pred - y_test)) / (2 * y_test.shape[0])
print('Mean Square Error for Linear Model is:', cost)
#remove 2 from the denominator to get the given value of cost
```

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
# Assuming x is a numpy array
n_rows = x_train.shape[0]  # Number of rows in x
sum_x = np.sum(x_train)  # Sum of all elements in x
squared_x = np.square(x_train)  # Square of x
sum_squared_x = np.sum(squared_x)  # Sum of square of x
sum_y = np.sum(y_train)  # Sum of all elements in y
sum_xy = np.dot(x_train, y_train)  # Dot product of x and y
sum_x_cube = np.sum(np.power(x_train,3))  # Sum of cube of x
sum_x_4degree = np.sum(np.power(x_train,4))  # Sum of 4th degree of x
sum_x_squared_y = np.dot(squared_x, y_train)  # Dot product of square of x and y

quad_matrixA = np.array([[n_rows, sum_x, sum_squared_x], [sum_x, sum_squared_x, sum_x_cube],
                    [sum_squared_x, sum_x_cube, sum_x_4degree]])  # Matrix of x
print('Matrix A:\n', quad_matrixA)

quad_matrixB = np.array([sum_y, sum_xy, sum_x_squared_y])  # Matrix of y

print('Matrix B:\n', quad_matrixB)
```
